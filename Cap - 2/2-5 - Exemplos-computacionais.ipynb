# CD-Notas

```python
import pandas as pd
notas = pd.read_excel('/kaggle/input/db-morettin/Dados_EB.xls', sheet_name='CD-Notas', skiprows=3)
```

### Figura 2.12 - Histograma

```python
import matplotlib.pyplot as plt

plt.hist(notas.stack(), bins=10)
plt.title('Histograma')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.show()
```

### Figura 2.13 - Dispersão Unidimensional

```python
from collections import Counter

contagem = Counter(notas.stack())

x = []
y = []

for valor, qtd in contagem.items():
    for i in range(qtd):
        x.append(valor)
        y.append(i)

plt.figure(figsize=(10,4))
plt.scatter(x, y, s=40)
plt.title('Gráfico de Dispersão Unidimensional')
plt.show()
```

### Figura 2.14 - Ramo-e-folha

```python
import stemgraphic

stemgraphic.stem_graphic(notas.stack())

plt.title('Gráfico Ramo-e-folha')
plt.show()
```

# CD-Poluicao

```python
polu = pd.read_excel('/kaggle/input/db-morettin/Dados_EB.xls', sheet_name='CD-Poluicao', skiprows=8)
```

### Figura 2.15 - Plot linear da série temporal

```python
plt.plot(polu['temp'])
plt.ylabel('Grau')
plt.xlabel('Dia')
```

### Figura 2.16 - Histograma

```python
plt.hist(polu['temp'])
plt.xlabel('Temperatura')
plt.show()
```

### Figura 2.17 - Gráfico de Disitribuição Unidimensional

```python
contagem = Counter(polu['temp'])

x = []
y = []

for valor, qtd in contagem.items():
    for i in range(qtd):
        x.append(valor)
        y.append(i)

plt.figure(figsize=(10,4))
plt.scatter(x, y, s=20)
plt.title('Gráfico de Dispersão Unidimensional')
plt.show()
```

```python
stemgraphic.stem_graphic(polu['temp'])

plt.title('Gráfico Ramo-e-folha')
plt.show()
```
